```python
from tensorflow.keras.models import load_model
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
model = load_model('model/cat_dog_model.h5')
```

```python
def image_predict(x):
    dt = cv2.imread(x)
    dt = cv2.resize(dt,(150,150))
    plt.imshow(dt)
    plt.show()
    dt = dt.reshape(1,150,150,3)
    result = np.where(model.predict(dt) >0.5,'강아지','고양이')[0][0]
    print(f'선택하신 사진은 {result} 입니다')
```

```python
image_predict('1.jpg')
```
