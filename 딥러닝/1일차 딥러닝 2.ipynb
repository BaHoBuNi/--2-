```python
def sigmoid(x):
    epsilon = 1e-5
    m = np.max(x)
    x = x/(m+epsilon)
    return 1/(1+np.exp(-x))
```

```python
def relu(x):
    return np.where(x > 0,x,0)
```

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(-x)+np.exp(x))
```
