### 식 설명 

df.best_month라는 배열에서 유일한 원소들과 그 빈도수를 구하고, 가장 빈도수가 높은 원소를 best_month라는 변수에 저장하는 코드

numpy.unique 함수는 배열에서 중복되지 않은 원소들을 정렬하여 반환하는 함수, return_counts 옵션을 True로 설정하면 각 원소의 개수도 함께 반환, 

np.argmax 함수는 배열에서 최대값의 위치를 반환하는 함수, 따라서 이 코드는 df.best_month에서 가장 많이 나타나는 원소의 위치를 찾아서 그 값을 best_month에 할당

#### duplciated 

 판다스 데이터프레임에서 중복된 행을 찾아내는 함수입니다.이 함수를 호출하면, 
 중복된 행이 있는 경우에는 해당 행이 True로 표시된 불리언 시리즈 객체를 반환

```python
df.duplicated()
```

```python
df.loc[5] = ['a',1,1]
```

```python
df
```

```python
df.duplicated(keep='last')

# 마지막 행 제외 True
```

```python
keep = 'first'
dup = []
for i in range(df.shape[0]-1):
    for j in range(df.shape[0]-i-1):
        if sum(df.iloc[i,:] == df.iloc[j+i+1,:]) == df.shape[1]:
            first = i
            last = j+i+1
            if keep == 'first':
                dup.append(last)
            else:
                dup.append(first)

dup = np.unique(dup)
result = np.full(df.shape[0],False)
result[dup] = True
df.loc[~result,:]
```

```python
# 중복 항목 지우기 함수화
```

```python
def my_duplicate(df,keep='first'):
    keep = 'first'
    dup = []
    for i in range(df.shape[0]-1):
        for j in range(df.shape[0]-i-1):
            if sum(df.iloc[i,:] == df.iloc[j+i+1,:]) == df.shape[1]:
                first = i
                last = j+i+1
                if keep == 'first':
                    dup.append(last)
                else:
                    dup.append(first)

    dup = np.unique(dup)
    result = np.full(df.shape[0],False)
    result[dup] = True
    return result
```

```python
my_duplicate(df)
```

```python
df.loc[my_duplicate(df),:]
```

```python
df.loc[~my_duplicate(df),:]
```

#### 정규화

정규화(Standardization 또는 Z-score normalization)는 데이터의 값의 스케일(scale)을 조정하는 작업

데이터의 스케일을 조정하면, 다른 스케일을 가진 변수들 간에 비교를 용이하게 할 수 있다. 또한, 데이터의 분포가 특정 범위 안에서 집중되도록 할 수 있다.

```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

#### 설명

StandardScaler는 데이터를 평균이 0, 분산이 1이 되도록 변환해주는 변환기(transformer) 클래스 이 클래스는 z-score 표준화 방법과 유사하다. 
이를 통해 변수 간의 스케일 차이가 크게 작용하지 않도록 데이터를 정규화할 수 있다.

### 전처리 feat 표준화, 정규화


StandardScaler와 MinMaxScaler는 둘 다 데이터 스케일링을 위한 전처리 방법 중 하나. 데이터 스케일링은 다양한 종류가 있으며, 가장 대표적인 것이 표준화(standardization)와 정규화(normalization)

StandardScaler는 표준화를 수행하는 전처리 방법으로, 각 피처(변수)의 평균과 표준편차를 사용하여 스케일링한다. 따라서, 스케일링된 데이터는 평균이 0, 표준편차가 1이 되도록 변환된다.

반면, MinMaxScaler는 정규화를 수행하는 전처리 방법으로, 각 피처(변수)의 최솟값과 최댓값을 사용하여 스케일링한다. 따라서, 스케일링된 데이터는 0과 1사이의 값으로 변환된다.

두 전처리 방법 모두 특정 분포를 따르는 데이터를 일정한 범위로 조정하는데 사용된다. 선택하는 전처리 방법은 데이터의 분포와 모델링 목적에 따라 다르게 선택되어야 한다

```python
sc = StandardScaler()
sc.fit(df)
```

정규화는 데이터셋의 특성을 바탕으로 스케일 조정을 하므로 데이터셋을 모두 다루어야한다. 따라서, 데이터프레임의 전체 행과 열을 이용하여 fit()을 호출

이 문제는 일반적으로 sklearn 모듈의 datasets 함수에서 데이터를 불러올 때 발생한다. datasets 함수는 딕셔너리 형태의 데이터를 반환하는데,

이 데이터를 바로 float() 함수에 넣으면 위와 같은 에러가 발생한다.

따라서, datasets 함수에서 반환된 딕셔너리 형태의 데이터를 pandas 또는 numpy와 같은 데이터 구조로 변환하여 사용하면 된다
